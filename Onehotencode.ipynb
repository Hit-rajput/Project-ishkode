```python
from numpy import asarray
```

```python
data = asarray([['red'], ['green'], ['blue']])
print(data)
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
one_encoder=OneHotEncoder(sparse=False)
Ordinal_encoder=OrdinalEncoder()
```

```python
ont_dt=one_encoder.fit_transform(data)
ordinal_dt=Ordinal_encoder.fit_transform(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data)
```

```python
print(ont_dt)
plt.hist(ont_dt)
```

```python
print(ordinal_dt)
plt.hist(ordinal_dt)
```
